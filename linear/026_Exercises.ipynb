## 1. Cross-validation linear regression

You need to change the variable ``alpha`` to be a list of alphas. Next do a loop and finally compare the results.

```python
import numpy as np

x = np.array([188, 181, 197, 168, 167, 187, 178, 194, 140, 176, 168, 192, 173, 142, 176]).reshape(-1, 1).reshape(15,1)
y = np.array([141, 106, 149, 59, 79, 136, 65, 136, 52, 87, 115, 140, 82, 69, 121]).reshape(-1, 1).reshape(15,1)

x = np.asmatrix(np.c_[np.ones((15,1)),x])

I = np.identity(2)
alphas = [0.01, 0.1, 1.0, 10.0]

for alpha in alphas:
    w = np.linalg.inv(x.T*x + alpha * I) * x.T * y
    w = np.asarray(w).ravel()
    print(f"alpha={alpha}: w={w}")
```

## 2. Implement based on the Ridge regression example, the Lasso regression.

Please implement the SGD method and compare the results with the sklearn Lasso regression results. 

```python
def sgd(x, y, alpha=0.1, learning_rate=1e-5, epochs=10, seed=42):
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float).ravel()

    n_samples, n_features = x.shape
    w = np.zeros(n_features, dtype=float)
    rng = np.random.default_rng(seed)

    for _ in range(epochs):
        for i in rng.permutation(n_samples):
            xi = x[i]
            error = xi @ w - y[i]

            grad = xi * error
            grad[1:] += alpha * np.sign(w[1:])

            w -= learning_rate * grad

    return w

```

```python
from sklearn.linear_model import Lasso

x = np.array([188, 181, 197, 168, 167, 187, 178, 194, 140, 176, 168, 192, 173, 142, 176]).reshape(-1, 1).reshape(15,1)
y = np.array([141, 106, 149, 59, 79, 136, 65, 136, 52, 87, 115, 140, 82, 69, 121]).reshape(-1, 1).reshape(15,1)

x = np.asmatrix(np.c_[np.ones((15,1)),x])

alpha = 0.1

w_sgd = sgd(x, y, alpha=alpha, learning_rate=1e-5, epochs=10, seed=42)
w_sgd = np.asarray(w_sgd).ravel()

lasso = Lasso(alpha=alpha, fit_intercept=False, max_iter=10000)
lasso.fit(np.asarray(x), np.asarray(y).ravel())
w_sklearn = np.asarray(lasso.coef_).ravel()

print('SGD Lasso weights:     ', w_sgd)
print('sklearn Lasso weights: ', w_sklearn)
# We used only 10 epochs thast's why the weights are so different.
```
